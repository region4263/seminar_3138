```python
df = data_frame_zoo[data_frame_zoo.animal == 'tiger']
s = df.sort_values(by=['water_need'])
graf = plt.plot(s.water_need)
```

```python
zebra_data = data_frame_zoo[data_frame_zoo.animal == 'zebra']
width = 0.35

fig, ax = plt.subplots()
rect = ax.bar(
    zebra_data.uniq_id, zebra_data.water_need, width, label='Зембра'
)

ax.set_ylabel('Условные единицы')
ax.set_xlabel('uniq_id')
ax.set_title('Потребность в воде')
ax.legend()

ax.bar_label(rect, padding=3)

fig.tight_layout()

plt.show()
```
